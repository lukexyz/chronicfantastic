# 🔴 Setup Room
**1.** Create `VDO.Ninja` director room  
**2.** Create joiner URL


Docs: https://docs.vdo.ninja/getting-started/the-power-of-the-url-parameter

# 1. `Create Room` using URL params

1. Create room as director using `?director` url parameter   
i.e. https://vdo.ninja/?director=chronicfantastic2000  



2. Guests use `?room` param
- Guests can see director, but not other guests videos
- Director can choose to perform as well

#### ⚙️ URL PARAMS ⚙️ [link](https://docs.vdo.ninja/advanced-settings/cheat-sheet-of-basic-parameters)
✅ Video only with `&view`  
✅ Show display names `&d`  
✅ Ask for display name `&l`  
✅ `&broadcast` video connections only from director, better performance

`&push` unique streamID, generated automatically if not assigned.

# 2. `Join URL` using URL params

```python
#| export
def get_joiner_url(roomname, username, session):
    # new users use this link
    return f'https://vdo.ninja/?room={roomname}&password={session}&broadcast&l={username}&m&ad&sl&view'

def get_base_url(roomname, session):
    # director initiates room with this link
    return f'https://vdo.ninja/?director={roomname}&password={session}'
```

```python
#| export
def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    # generate random session id
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
id_generator(3)
```

```python
roomname = 'chronicfantastic2000'
director = "director"
username = 'luke'
session = id_generator(3)

baseurl = get_base_url(roomname, session)
joiner_url = get_joiner_url(roomname, username, session)

display(Markdown(f'**`Director`**</br> 🔗 {baseurl}'))
display(Markdown(f'**`Joiner_URL`**</br> 🔗 {joiner_url}'))
```
